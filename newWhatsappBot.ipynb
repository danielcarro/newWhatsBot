```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import urllib
import random

# Mensagem fixa para todos os contatos
MENSAGEM = """
VocÃª tem conhecimento em alguma Ã¡rea e quer transformar isso em uma fonte de renda? Com o *SABER++*, criar e vender cursos online nunca foi tÃ£o fÃ¡cil! ðŸŽ“ðŸ’°  

âœ… Crie cursos profissionais sem precisar de experiÃªncia tÃ©cnica  
âœ… Venda seus cursos de forma automatizada  
âœ… Gerencie alunos, pagamentos e certificados em um sÃ³ lugar  
âœ… Utilize InteligÃªncia Artificial para otimizar seu conteÃºdo  

ðŸš€ DÃª o primeiro passo para o seu sucesso digital! Comece agora com o *SABER++* e transforme seu conhecimento em um negÃ³cio lucrativo.  

ðŸ”— https://www.saberapps.com.br

#Empreendedorismo #CursosOnline #GanharDinheiro #SaberPlusPlus #NegÃ³cioDigital
"""

# FunÃ§Ã£o para enviar mensagem
def enviar_mensagem(navegador, pessoa, numero, mensagem):
    try:
        texto = urllib.parse.quote(mensagem)
        link = f"https://web.whatsapp.com/send?phone={numero}&text={texto}"
        navegador.get(link)

        # Espera o campo de mensagem carregar
        campo_mensagem = WebDriverWait(navegador, 30).until(
            EC.presence_of_element_located((By.XPATH, '//div[@contenteditable="true"][@data-tab="10"]'))
        )

        # Foca no campo de mensagem
        campo_mensagem.click()

        # Espera o botÃ£o de enviar ficar clicÃ¡vel
        botao_enviar = WebDriverWait(navegador, 30).until(
            EC.element_to_be_clickable((By.XPATH, '//span[@data-icon="send"]'))
        )

        # Adiciona um pequeno atraso antes de clicar
        time.sleep(2)

        # Usa JavaScript para clicar no botÃ£o
        navegador.execute_script("arguments[0].click();", botao_enviar)
        print(f"Mensagem enviada para {pessoa} ({numero})")

    except Exception as e:
        print(f"Erro ao enviar mensagem para {pessoa} ({numero}): {e}")

# FunÃ§Ã£o principal
def main():
    # Carrega os contatos da planilha
    contatos_df = pd.read_excel("contatos.xlsx")

    # Configura o WebDriver
    chrome_driver_path = r'C:\Users\objet\AppData\Local\Programs\Python\Python39\chromedriver.exe'  # Atualize este caminho
    service = Service(chrome_driver_path)

    chrome_options = Options()
    chrome_options.add_argument("--disable-extensions")
    chrome_options.add_argument("--disable-gpu")
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-dev-shm-usage")
    chrome_options.add_argument("--start-maximized")

    navegador = webdriver.Chrome(service=service, options=chrome_options)

    # Abre o WhatsApp Web
    navegador.get("https://web.whatsapp.com/")
    print("Escaneie o QR Code para continuar...")

    try:
        WebDriverWait(navegador, 60).until(EC.presence_of_element_located((By.ID, "side")))
        print("Login no WhatsApp Web realizado com sucesso!")
    except Exception as e:
        print(f"Erro ao carregar WhatsApp Web: {e}")
        navegador.quit()
        return

    # Envia a mensagem para todos os contatos
    for i, contato in contatos_df.iterrows():
        pessoa = contato["Pessoa"]
        numero = str(contato["NÃºmero"]).strip()

        if not pessoa or not numero:
            print(f"Dados incompletos para {pessoa} ({numero}). Pulando...")
            continue

        enviar_mensagem(navegador, pessoa, numero, MENSAGEM)
        tempo_espera = random.randint(10, 20)
        time.sleep(tempo_espera)

    # Fecha o navegador ao final
    navegador.quit()
    print("Envio de mensagens concluÃ­do.")

if __name__ == "__main__":
    main()
```
